### STEP 1: Import Python Library and Read CSV Kagle Dataset


```python
import numpy as np
import pandas as pd

# Load the dataset
data = pd.read_csv('D:/TheCoder/DataScience/DS_SupervisedMachineLearning/datasets/AmesHousing.csv')
datas = data.head(3)
print(datas)

#Step 1: Calculate the means of x and y
x = data['Gr Liv Area'].values
y = data['SalePrice'].values
```

### STEP 2: Calculate The Mean X and Y


```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(f'X mean {x_mean}')
print(f'Y mean {y_mean}')
```

### STEP 3: Calculate b1 (Slope) and Intercept (b0)


```python
#b1 = E (Xi - Xmean) * (Yi - Ymean) / pow(Xi - Xmean)
numerator = 0
denominator = 0
for i in range(len(x)):
    numerator += (x[i] - x_mean) * (y[i] - y_mean)
    denominator += (x[i] - x_mean) ** 2
b1 = numerator / denominator
b0 = y_mean - (b1 * x_mean)
print(f'Slope (b1):{b1}')
print(f'Intercept (b0): {b0}')
```

### STEP 4: Create Model (Prediction for Incoming X)


```python
y_pred = b0 + b1 * x
print(y_pred)
```

### STEP 5: Calculate The Mean Squared Error (MSE)


```python
mse = np.mean(y - y_pred ** 2)
ss_tot = np.sum((y - y_mean) ** 2)
ss_res = np.sum((y - y_pred) ** 2)
r2_score = 1 - (ss_res / ss_tot)
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2_score}')
```

### STEP 5: Visualize The Model


```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='green', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Gr Live Area')
plt.ylabel('SalePrice')
plt.legend()
plt.show
```
